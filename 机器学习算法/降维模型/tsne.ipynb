# t-SNE（t 分布随机邻域嵌入）
是一种非线性降维技术，常用于高维数据的可视化。t-SNNE 的主要思想是：
- 首先，t-SNE 使用高斯分布将高维数据映射到低维空间中。
- 然后，t-SNE 使用概率分布的思想，将高维数据点之间的相似度映射到低维空间中。
- 最后，t-SNE 使用随机游走算法，在低维空间中生成连续的分布。

t-SNE 的优点：
- 它可以有效地将高维数据映射到二维或三维空间中，使得数据点之间的距离变得更加直观。
- 它可以保留原始数据点之间的结构信息，因此可以保留原始数据点之间的相似性。
- 它可以发现数据中的聚类结构，并用不同的颜色或符号表示不同的类别。

t-SNE 的缺点：
- 它需要大量的计算资源，尤其是当数据量较大时。
- 它对初始条件敏感，初始条件的选择会影响结果的精度。

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.manifold import TSNE

# 1. 生成示例高维数据
# 使用 make_blobs 函数生成 100 个样本，特征维度为 10，分为 3 个簇的数据集
X, y = make_blobs(n_samples=100, n_features=10, centers=3, random_state=42)

# 2. 创建 t-SNE 模型并进行降维
# 创建 t-SNE 模型，设置降维后的维度为 2，学习率为 200，随机种子为 42
tsne = TSNE(n_components=2, learning_rate=200, random_state=42)
# 对数据进行降维处理
X_tsne = tsne.fit_transform(X)

# 3. 可视化降维结果
plt.figure(figsize=(8, 6))
# 根据不同的簇标签使用不同颜色绘制散点图
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', s=50)
plt.title('t-SNE Dimensionality Reduction')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.colorbar()
plt.show()
```
