# 谱聚类
'''基于图论的聚类方法，通过对数据点之间的相似度矩阵进行特征分解来实现聚类。'''

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import SpectralClustering

# 生成示例数据
# make_moons 可以生成半月形的数据，适合展示谱聚类处理非凸形状簇的能力
X, y = make_moons(n_samples=200, noise=0.05, random_state=42)

# 创建谱聚类模型
# n_clusters 表示要划分的簇的数量，affinity 表示相似度矩阵的计算方式，这里使用 'nearest_neighbors'
clustering = SpectralClustering(n_clusters=2, affinity='nearest_neighbors')

# 进行聚类
clustering.fit(X)

# 获取聚类标签
labels = clustering.labels_

# 可视化结果
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Spectral Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
